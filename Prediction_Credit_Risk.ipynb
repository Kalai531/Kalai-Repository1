```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# --- Re-load and Process Data (Necessary setup for a standalone script) ---

# 1. Load the Dataset
df = pd.read_csv(r"C:/Users/hp/Downloads/Credit_Risk.csv")

# 5. Remove any duplicate rows
df.drop_duplicates(inplace=True)

# 6. Create 'AgeGroup' column
bins = [0, 30, 55, np.inf]
labels = ['Young', 'Middle-Aged', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# ------------------------------------------------------------------------

print("\n" + "="*50)
print("## 1. Check Data Types (df.dtypes) üîç")
# 1. Check Data Types
print(df.dtypes)

# ---
print("\n" + "="*50)
print("## 2. Descriptive Summary Statistics (df.describe()) üìä")
# 2. Summary Statistics for numerical columns
# Transpose for better screen readability
print(df.describe().transpose())

# ---
print("\n" + "="*50)
print("## 3. Correlation Matrix of Numerical Features ü§ù")
# 3. Correlation Matrix
numerical_df = df.select_dtypes(include=[np.number])
correlation_matrix = numerical_df.corr()

# Print the correlation matrix
print(correlation_matrix)

# Optional: Visualize the correlation matrix (requires matplotlib and seaborn)
# This is often the visual representation of the 'Expected Output'
print("\n")
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# ---

# ------------------------------------------------------------------------
```
